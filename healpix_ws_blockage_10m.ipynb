```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import healpy as hp
```

```python
NSIDE = 4096
print(
    "Approximate resolution at NSIDE {} is {:.2} deg".format(
        NSIDE, hp.nside2resol(NSIDE, arcmin=True) / 60
    )
)
```

```python
NPIX = hp.nside2npix(NSIDE)
print(NPIX)
```

```python
ori = np.loadtxt("./earth_10Mhz/B_10M_original.txt")
```

```python
hp.mollview(ori, nest=True, title="Mollview image B with NSIDE=4096")
hp.graticule()  # 添加经纬线
plt.show()
```

```python
total_data = np.loadtxt("./wsblockage10M/C1day1M.txt")

hp.mollview(total_data, nest=True, title="Mollview image B with NSIDE=4096 in one day")
hp.graticule()  # 添加经纬线
plt.show()
```
