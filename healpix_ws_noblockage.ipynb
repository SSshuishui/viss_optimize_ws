```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import healpy as hp
```

```python
NSIDE = 512
print(
    "Approximate resolution at NSIDE {} is {:.2} deg".format(
        NSIDE, hp.nside2resol(NSIDE, arcmin=True) / 60
    )
)
```

```python
NPIX = hp.nside2npix(NSIDE)
print(NPIX)
```

```python
ori = np.loadtxt("./earth_1Mhz/B.txt")
```

```python
hp.mollview(ori, nest=True, title="Mollview image B with NSIDE=512")
hp.graticule()  # 添加经纬线
plt.show()
```

```python
total_data = np.zeros_like(np.loadtxt("./3dnoblockage/C1day1M.txt"))

# 读取并累加30个文件
for i in range(1, 31):
    file_path = f"./3dnoblockage/C{i}day1M.txt"
    data = np.loadtxt(file_path)
    total_data += data

hp.mollview(total_data, nest=True, title="Mollview image B with NSIDE=512")
hp.graticule()  # 添加经纬线
plt.show()
```
